## Subjective Feedbacks

Some comments provided by the participants over the accompanniment system are:

> The melodies are very similar, and it is difficult to identify the specific melody.  
> The accompaniments for the higher part of the melody is better, but the transitions between music phrases are a bit strange.  
> There are some rhythms not fitted well enough.  
> It's easy to be monotonous, and some pieces have no emotion in them.  
> The generated melodies lack variations and complexity.  
> Only follows the basic melody, is not appriciate enough.  
> The automatic generated accompaniment can fit the given melody, however, the musical quality doesn’t sound well sometimes.  
> For non-composers, if his expectation is not very high, the automatic accompaniment can give a reasonable starting point.  
> Some chords are not harmony at all, sounds like a child playing the piano full of mistakes.  
> Some tunes sounds very messy, which doesn’t feel like it’s a united composition.  
> It should follow the basic rules of composing music. So far I feel it might be very useful in electronic music.  
> It makes sense to some extent. Although it quite easy to recognize, at least the main rhythm is captured.  
> Some pieces are noisy.  
> Sounds too flat, lacking of some diversities to inflame the listener’s passion.  
> Can give a starting point for non-composers, but couldn’t get the non-composers to an advanced stage.  
> The accompaniment has too narrow range and too slow pace compared to the melody.  
> Some principles of generating the accompaniment are worth learning.  
> The music pieces sounded not natural and harmonious enough.  
> The music segments don’t have a major pattern in it. For example, in a real song, there always exists some pattern in it. Mostly, a folk song only needs 4 or 5 chords through all of the song. The generated music assembles the combination of notes.  
> The endings of the music is a bit unexpected, a lot note repetitions in the final piece. But the first piece for this quesiton is pretty good.  
> Not stable enough in the transitions. Some mistakes in the harmonicity and key following. Not giving a clear ending for music phrases.  
> People will get some ideas from the piano roll.
